# Contoh perintah Python Tuples

Tuple digunakan untuk menyimpan beberapa item dalam satu variabel.

Tuple adalah salah satu dari 4 tipe data bawaan dalam Python yang digunakan untuk menyimpan kumpulan data, 3 lainnya adalah Daftar, Set, dan Kamus, semuanya dengan kualitas dan penggunaan yang berbeda.

**Tuple adalah kumpulan yang dipesan dan tidak dapat diubah.**

Tupel ditulis dengan tanda kurung bulat.

```python
# mendefinisikan isi tuples
thistuple = ("apple", "banana", "cherry", "apple", "cherry")
print(thistuple)
```

```python
# untuk mengetahui panjang tau banyaknya itemps pada tuples
thistuple = ("apple", "banana", "cherry")
print(len(thistuple))
```

```python
#Untuk membuat Tuple dengan hanya satu item, Anda harus menambahkan koma setelah item, jika tidak Python tidak akan mengenalinya sebagai Tuple.

thistuple = ("apple",) # harus menambahkan koma jika satu items agar terihat bahwa tuples
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
```

```python
tuple1 = ("apple", "banana", "cherry") #type string menjadi tuple
tuple2 = (1, 5, 7, 9, 3) #type integer menjadi tuple
tuple3 = (True, False, False) #type boolean menjadi tuple
print(type(tuple1))
print(type(tuple2))
print(type(tuple3))
```

```python
tuple1 = ("abc", 34, True, 40, "male") #bisa di gabungin beberapa type data
print(type(tuple1))
```

```python
mytuple = ("apple", "banana", "cherry")
print(type(mytuple))
```

```python
thistuple = tuple(("apple", "banana", "cherry"))# perhatikan tanda kurung bulat ganda
print(thistuple)
```

```python
# Anda dapat menentukan rentang indeks dengan menentukan di mana harus memulai dan di mana harus mengakhiri rentang.
# Saat menentukan rentang, nilai yang dikembalikan akan menjadi tupel baru dengan item yang ditentukan.
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])
print(thistuple[:4])
print(thistuple[2:])
print(thistuple[-4:-1])
```

```python
# Periksa apakah "apple" ada di Tuple:
thistuple = ("apple", "banana", "cherry")
if "apple" in thistuple:
  print("Yes, 'apple' is in the fruits tuple")
```

```python
#Setelah tuple dibuat, Anda tidak dapat mengubah nilainya. Tuple tidak dapat diubah, atau tidak dapat diubah seperti yang juga disebut.
#Tapi ada solusi. Anda dapat mengonversi Tuple menjadi daftar, mengubah daftar, dan mengonversi daftar kembali menjadi Tuple.
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

```python
# Add items
# Karena tupel tidak dapat diubah, mereka tidak memiliki metode append() bawaan, tetapi ada cara lain untuk menambahkan item ke tupel.
# 1. Convert into a list: Sama seperti solusi untuk mengubah tupel, Anda dapat mengubahnya menjadi daftar, 
#    menambahkan item Anda, dan mengubahnya kembali menjadi tupel.thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
```

```python
# 2. Tambahkan tupel ke tupel. Anda diperbolehkan untuk menambahkan tupel ke tupel, jadi jika Anda ingin menambahkan satu item, (atau banyak), 
#    buat tupel baru dengan item tersebut, dan tambahkan ke tupel yang ada:
thistuple = ("apple", "banana", "cherry")
y = ("orange",)
thistuple += y

print(thistuple)
```

```python
# Remove Items
# Tuple tidak dapat diubah, jadi Anda tidak dapat menghapus item darinya, 
# tetapi Anda dapat menggunakan solusi yang sama seperti yang kami gunakan untuk mengubah dan menambahkan item Tuple:
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
print(thistuple)
```

```python
Kata key del dapat menghapus Tuple sepenuhnya:
thistuple = ("apple", "banana", "cherry")
del thistuple
print(thistuple) #ini akan menimbulkan kesalahan karena tupel tidak ada lagi/akan error
```

```python
# Membuka Tuple
# Ketika kita membuat sebuah tuple, kita biasanya memberikan Value padanya. Ini disebut "mengemas" sebuah tuple:
fruits = ("apple", "banana", "cherry")

(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

```python
# Menggunakan Asterisk (*)
# Jika jumlah variabel kurang dari jumlah nilai, Anda dapat menambahkan * ke nama variabel dan nilai akan ditetapkan ke variabel sebagai daftar:
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)
```

```python
# Contoh
# Tambahkan daftar nilai variabel "tropic":
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

```python
# Loop Melalui Tuple
# Anda dapat mengulang item Tuple dengan menggunakan for loop.
thistuple = ("apple", "banana", "cherry")
for x in thistuple:
  print(x)
```

```python
# Loop Nomor Indeks
# Anda juga dapat loop item Tuple dengan mengacu pada nomor indeksnya.
# Gunakan fungsi range() dan len() untuk membuat iterable yang sesuai.
thistuple = ("apple", "banana", "cherry")
for i in range(len(thistuple)):
  print(thistuple[i])
```

```python
# Menggunakan while loop
# Anda dapat mengulang item list dengan menggunakan loop while.
# Gunakan fungsi len() untuk menentukan panjang tuple, lalu mulai dari 0 dan ulangi item tuple dengan mengacu pada indeksnya.
thistuple = ("apple", "banana", "cherry")
i = 0
while i < len(thistuple):
  print(thistuple[i])
  i = i + 1
```

```python
# Join Two Tuples
# To join two or more tuples you can use the + operator:
tuple1 = ("a", "b" , "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
# Multiply Tuples
# Jika Anda ingin mengalikan konten Tuple beberapa kali, Anda dapat menggunakan operator *:
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2

print(mytuple)
```

```python
# Metode update() akan memperbarui kamus dengan item dari argumen yang diberikan.
# Argumen harus berupa kamus, atau objek yang dapat diubah dengan pasangan kunci:nilai.
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"year": 2020})
```

```python
# Menambahkan item ke kamus dilakukan dengan menggunakan kunci indeks baru dan menetapkan nilai untuk itu:
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red"
print(thisdict)
```

```python
# Metode update() akan kamus dengan item dari argumen yang diberikan. Jika item tidak ada, item akan ditambahkan.
# Argumen harus berupa kamus, atau objek yang dapat diubah dengan pasangan kunci:nilai.
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"color": "red"})
```

```python
# Ada beberapa metode untuk menghapus item dari kamus:
# Metode pop() menghapus item dengan nama kunci yang ditentukan:
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

```python
# Metode popitem() menghapus item yang terakhir dimasukkan :
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
print(thisdict)
```

```python
# Kata kunci del menghapus item dengan nama kunci yang ditentukan:
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"]
print(thisdict)
```

```python
# Kata kunci del juga dapat menghapus kamus sepenuhnya:
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict
print(thisdict) #ini akan menyebabkan kesalahan karena "thisdict" tidak ada lagi. Hasil nya akan error
```

```python
#The clea r() method empties the dictionary:

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.clear()
print(thisdict)
```

```python
# Loop Dictionary
# Anda dapat mengulang kamus dengan menggunakan for loop.
for x in thisdict:
  print(x)
```

```python
# Print all values in the dictionary, one by one:

for x in thisdict:
  print(thisdict[x])
```

```python
# Anda juga dapat menggunakan metode values() untuk mengembalikan nilai kamus:

for x in thisdict.values():
  print(x)
```

```python
# Anda dapat menggunakan metode keys() untuk mengembalikan kunci kamus:

for x in thisdict.keys():
  print(x)
```

```python
# Loop through both keys and values, by using the items() method:

for x, y in thisdict.items():
  print(x, y)
```

```python
# Copy Dictionary
# Anda tidak dapat menyalin kamus hanya dengan mengetikkan dict2 = dict1, karena: dict2 hanya akan menjadi referensi ke dict1, 
# dan perubahan yang dibuat di dict1 akan otomatis juga dibuat di dict2.
# Ada beberapa cara untuk membuat salinan, salah satu caranya adalah dengan menggunakan metode Kamus bawaan copy().
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = thisdict.copy()
print(mydict)
```

```python
# Cara lain untuk membuat salinan adalah dengan menggunakan fungsi bawaan dict().
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = dict(thisdict)
print(mydict)
```

```python
# Kamus Bersarang
# Kamus dapat berisi kamus, ini disebut kamus bersarang.
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
print(myfamily)
```

```python
# Atau, jika Anda ingin menambahkan tiga kamus ke dalam kamus baru:
# Buat tiga kamus, lalu buat satu kamus yang akan berisi tiga kamus lainnya:
child1 = {
  "name" : "Emil",
  "year" : 2004
}
child2 = {
  "name" : "Tobias",
  "year" : 2007
}
child3 = {
  "name" : "Linus",
  "year" : 2011
}

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}
print(myfamily)
```
