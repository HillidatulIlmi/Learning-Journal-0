## Anaconda Basic

Anaconda adalah distribusi bahasa pemrograman Python dan R untuk komputasi ilmiah, yang bertujuan untuk menyederhanakan manajemen dan penerapan paket. Distribusinya mencakup paket ilmu data yang cocok untuk Windows, Linux, dan macOS.

Download Anaconda from [here](https://www.anaconda.com/products/individual).



### Managing Conda

Verifikasi bahwa conda diinstal dan berjalan di sistem Anda dengan mengetik di CMD:

`conda --version`

Conda menampilkan nomor versi yang telah Anda instal. Anda tidak perlu menavigasi ke direktori Anaconda.

Contoh: `conda 4.7.12`

Perbarui conda ke versi saat ini. Ketik berikut ini :

`conda update conda`

Conda membandingkan versi dan kemudian menampilkan apa yang tersedia untuk diinstal.

Jika versi conda yang lebih baru tersedia, ketik y untuk memperbarui:

`Proceed ([y]/n)? y`

### Managing Environtment

Conda memungkinkan Anda membuat environments terpisah yang berisi file, paket, dan dependensinya yang tidak akan berinteraksi dengan environments lain.

Saat Anda mulai menggunakan conda, Anda sudah memiliki environments default bernama base. Anda tidak ingin memasukkan program ke dalam environments dasar Anda. Buat environments terpisah untuk menjaga program Anda tetap terisolasi satu sama lain.

Buat environments baru dan instal paket di dalamnya.

Kami akan memberi nama environments `Hacktiv8` dan menginstal paket pandas. Di Anaconda Prompt atau di jendela terminal Anda, ketikkan yang berikut ini:

`conda create --name Hacktiv8 pandas`

Conda memeriksa untuk melihat paket tambahan ("dependensi") apa yang dibutuhkan panda, dan menanyakan apakah Anda ingin melanjutkan :

`Proceed (y]/n)? y`

Ketik "y" dan tekan Enter untuk melanjutkan.

Untuk menggunakan, atau "mengaktifkan" environments baru, ketik berikut ini:

```sh
Windows: conda activate Hacktiv8

macOS and Linux: conda activate Hacktiv8
```

Sekarang Anda berada di environments Hacktiv8 Anda, semua perintah conda yang Anda ketik akan masuk ke environments itu sampai Anda menonaktifkannya.

Untuk melihat daftar semua environments Anda, ketik :

`conda info --envs`

Daftar environments muncul, mirip dengan berikut ini:

```sh
conda environments:

    base           /home/username/Anaconda3
    snowflakes   * /home/username/Anaconda3/envs/snowflakes
```

environments aktif adalah environments dengan tanda bintang (*).

Ubah environments Anda saat ini kembali ke default (basis): `conda activate`

### Managing Python

Saat Anda membuat environments baru, conda menginstal versi Python yang sama dengan yang Anda gunakan saat mengunduh dan menginstal Anaconda. Jika Anda ingin menggunakan versi Python yang berbeda, misalnya Python 3.6, cukup buat environments baru dan tentukan versi Python yang Anda inginkan.

Buat lingkungan baru bernama "fox" yang berisi Python 3.6:

`conda create --name fox python=3.6`

Ketika conda bertanya apakah Anda ingin melanjutkan, ketik "y" dan tekan Enter.

### Managing Packages

Di bagian ini, Anda memeriksa paket mana yang telah Anda instal, memeriksa mana yang tersedia dan mencari paket tertentu dan menginstalnya.

Untuk menemukan paket yang telah Anda instal, aktifkan terlebih dahulu lingkungan yang ingin Anda cari. Lihat di atas untuk perintah untuk mengaktifkan lingkungan Hacktiv8 Anda.

Periksa untuk melihat apakah paket yang belum Anda instal bernama "beautifulsoup4" tersedia dari repositori Anaconda (harus terhubung ke Internet):

`conda search beautifulsoup4`

Conda menampilkan daftar semua paket dengan nama itu di repositori Anaconda, jadi kami tahu itu tersedia.

Instal paket ini ke lingkungan saat ini:

`conda install beautifulsoup4`

Periksa untuk melihat apakah program yang baru diinstal ada di lingkungan ini:
`conda list`
