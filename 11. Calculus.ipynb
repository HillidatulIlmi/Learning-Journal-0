# Optimisasi

Optimization on Code//Find Minimum. Secara turunan numerik, tidak menganggap input atau output sebagai simbol namun hasil sampel berupa data/array

```python
from scipy.optimize import minimize_scalar

def f(x):
  return 2*x**2+4*x-1

optimisasi =minimize_scalar(f)
optimisasi
```

partial derivative adalah mengukur nilai y yang berubah karena x dan z

```python
def f2(x):
  return -(x**3 - 6*x**2 + 9*x + 1)

optimisasi = minimize_scalar(f2)
optimisasi
```

### Integral Numerical-Function

fungsi soalnya -> f(x) = 4x^3 - 2x^2 - 4, dengan batas bawah adaah 2 dan batas atas adalah 5

```python
def integral_f(x):
  return 4*x**3 + 2*x**2 - x

#Integral tertentu dari f(x) dari 2 - 5
integral_f(5) - integral_f(2)
```

```python
from scipy import integrate

def f(x):
  return 4*x**3 + 2*x**2 - x
  
I = integrate.quad(f,2,5)
print(I)
```

### Luas Area Lingkaran

```python
from scipy.integrate import trapz
import numpy as np

R=30
x=np.linspace(-R,R,100000)

upper = np.sqrt(R**2 - x**2)
lower = -upper
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(6,6))
ax.plot(x,upper)
ax.plot(x,lower)
```

```python
area = trapz(upper,x) - trapz(lower,x)

print(area)
```

```python
def upper(x):
  return np.sqrt(R**2 - x**2)

def lower(x):
  return -np.sqrt(R**2 - x**2)

area = integrate.quad(upper,-R,R)[0] - integrate.quad(lower,-R,R)[0]
print(area)
```
