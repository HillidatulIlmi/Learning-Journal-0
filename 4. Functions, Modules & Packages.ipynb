### Fungsi dengan argumen

```python
def luas_segitiga(alas, tinggi):
    print(1/2 * alas * tinggi)

luas_segitiga(2,3) + 3

#akan error karena fungsi hanya berisi print bukan return
```

```python
def luas_segitiga(alas, tinggi):
    return (1/2 * alas * tinggi)

luas_segitiga(2,3) + 3
```

```python
def luas_persegi(sisi = 3):
    luas = sisi*sisi
    return luas

luas_persegi(2)
luas_persegi()
```

## *ARGS dan **KWARGS

```python
def my_function(*nama):
    print("Teman saya adalah " + nama[2])

my_function("Emil", "Tobias", "Linus")
```

```python
def my_function(nama3, nama2, nama1):
    print("nama teman saya adalah " + nama3)

my_function(nama1 ="Emil", nama2 ="Tobias", nama3 ="Linus")
```

```python
def my_function1(**nama):
    print('nama dia adalah', nama['nama1'])

my_function1(nama1 = 'linus', nama2 = 'tobias')
```

## default parameter value

```python
def my_country(country = 'Indonesia'):
    print('I am from', country)
    
my_country()
my_country('Swiss')
```

```python
x = 'luar biasa'

def myfunc():
    x = 'pintar'
    print('siswa ini', x)

# myfunc()
print('saya ', x)
myfunc()
```

## Lambda Function

```python
## tanpa lambda
def square_sum(x,y):
    return (x**2 + y**2)

square = lambda x,y : x**2 + y**2
square(2,2)
```

```python
greeting = lambda name : print(f"Hello, {name}")
sapa = greeting
sapa('hana')
```
