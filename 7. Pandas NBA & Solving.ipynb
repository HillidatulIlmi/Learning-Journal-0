## **Columns Manipulation**

- rename column
- drop column
- create new column

```python
nba = df.copy()
```

```python
nba.head()
```

```python
# rename
nba.rename(columns={'game_location': 'loc', 'game_result':'result'}, inplace=True)
```

```python
nba
```

```python
# drop column
nba.drop(labels=['elo_i', 'elo_n', 'opp_elo_i', 'opp_elo_n'], axis=1, inplace=True)
```

```python
nba
```

```python
# create new column
nba.head()
```

```python
nba['selisih'] = nba['pts'] - nba['opp_pts']
```

```python
nba
```

```python
nba.columns
```

## latihan

```python
nba.head()
```

```python
# siapkan list kosong
# definisikan argumen dari for loop
# simpan hasil ke dalam list kosong
# jika sudah selesai convert list menjadi kolom

# kolom pts -> nba.iloc[:, 10]
# kolom opp_pts -> nba.iloc[:, 14]

hasil = []

for idx in range(0,len(nba)):
    selisih = nba.iloc[idx, 10] - nba.iloc[idx, 14]
    hasil.append(selisih)

hasil
```

```python
nba['selisih_loop'] = hasil
```
