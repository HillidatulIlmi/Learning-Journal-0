# Distribusi Normal

```python
#Melihat bentuk umum atau visualisasi sebuah distribusi normal
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

#PDF Normal
x = np.linspace(-3, 3)
p = norm.pdf(x, 0, 1) #pdf -> probability distribution function

plt.plot(x, p, 'k'); # k itu warna garisnya hitam
#sebuah seberan data yang dibuat histogtam
```

**Distribusi Normal Baku**

Rumus nya  -> 
Z = (x-miu)/sigma

x = nilai dari data

miu = mean (rata-rata)

sigma = standar deviasi (simpangan baku)

```python
#Distribusi Normal baku
data = np.random.normal(170, 10, 250)
 #170 itu rata-rat yang garis tengahnya berada di 170
 # 10 itu standar deviasi atau seberan datanya
 # 250 adalah jumlah datanya
```

```python
# dilihat bahwa mean 170 ada di tengah
data.mean()
```

```python
data.std()
```

```python
#Histogram distribusi normal baku
plt.hist(data, bins=25)
```

```python
#Distribusi normal dengan data dari plot sebelumnya
mu = data.mean()
std = data.std()

#Histogram dari data histogram sebelumnya
plt.hist(data, bins=25, density = True); # density itu melakukan bentuk data random yang kita buat seolah-olah berdistribusi normal
xmin, xmax = plt.xlim()  #Diambil nilai maksimum dan minimumnya
#linspace itu metode untuk mengrange data dari 0-1 sebanyak 100
x = np.linspace(xmin, xmax, 100) # sumbu x
p = norm.pdf(x, mu, std) # sumbu y
#Baru di plot pdf nya
plt.plot(x, p, 'k', linewidth=2) #linewidth itu garis yang membentuk distribusinya
```

Nilai di sekitaran 145 sampai 150 tidak ada datanya, tetapi karena dihampiri data yang berdistribusi normal ada nilai likelihood nya makanya masih membentuk garis. likelihood sebagai sumbu x nya
