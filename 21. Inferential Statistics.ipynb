### Pengujian Hipotesis Sampel Tunggal/Single sample one sided

Mari kita berperan. Misalkan selama lima tahun terakhir, pendapatan harian kita rata-rata \$500 dan selama setahun terakhir, kita mencapai rata-rata \$580 sehari. Apakah itu berarti bahwa kita meningkat secara signifikan?

```python
data['income'] = data['Quantity']*data['UnitPrice'] #variabel data income berisikan data quantity control pada barang unitprice
data['date'] = data['InvoiceDate'].dt.date #variabel data berisikan tahun dilakukannya produksi
daily_income = data[['date','income']].groupby('date').sum() #melihat daily_income dari dari data jumlah quantity control pada waktu 
print('Average Income a Day for the last a year: ${}'.format(np.round(daily_income['income'].mean()))) #melihat rata-rata
```

Untuk memeriksa apakah penjualan kami meningkat secara signifikan atau tidak, kami akan melakukan sampel tunggal satu sisi dan menetapkan tingkat signifikansi 0,05. Kami menggunakan metode ini karena kami hanya menguji variabel dan membandingkan sampel (data satu tahun terakhir) dan populasi (kami menganggap itu adalah data lima tahun terakhir).

Our hypothesis on this case:

**H0: μ <= \$500**

**H1: μ > \$500**

```python
#variabel daily_income 
daily_income
```

```python
daily_income.mean()
```

```python
#ingin melihat apakah 500 dolar itu signifikan
#stats.ttest_1samp syntax untuk one sample t-test
#Membandigkan daily_income dengan 500 sebagai patokannya
t_stat,p_val = stats.ttest_1samp(daily_income.income, 500)
print('P-value:',p_val/2) #Nilai p dibagi 2 karena outputnya adalah nilai p dua sisi
print('t-statistics:',t_stat)
```

Hasil 0,002 lebih kecil dari 0,05 maka dengan ini menolak h0. yang artinya ada perbedaan

```python
daily_income_pop = np.random.normal(daily_income.income.mean(), daily_income.income.std(), 10000)

ci = stats.norm.interval(0.90, daily_income.income.mean(), daily_income.income.std())

plt.figure(figsize=(16,5))
sns.distplot(daily_income_pop, label='Daily Income (Populasi)', color='blue')
plt.axvline(daily_income.income.mean(), color='red', linewidth=2, label='Daily Income (Rata-rata)')
plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%/Tingkat Kepercayaan 0,05')
plt.axvline(daily_income_pop.mean() + t_stat*daily_income_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis/0.002')
plt.legend()
```

**Karena nilai p kami kurang dari `0,05`, jadi kami menolak hipotesis nol** dan kami dapat menyimpulkan bahwa, penjualan kami selama setahun terakhir meningkat daripada lima tahun terakhir.

*Catatan: Variabel `t-statistics` mengacu pada seberapa jauh hipotesis alternatif dari hipotesis nol.*

### One Sample Two Tailed

Our hypothesis on this case:

**H0: μ =\$500**

**H1: μ != \$500**

```python
t_stat,p_val = stats.ttest_1samp(daily_income.income, 500)
print('P-value:',p_val) #Tetap akan menolak h0 karena 0,005 < 0.05
print('t-statistics:',t_stat)
```

```python
daily_income_pop = np.random.normal(daily_income.income.mean(), daily_income.income.std(), 10000)

ci = stats.norm.interval(0.95, daily_income.income.mean(), daily_income.income.std()) #tingkat kepercayaannya

plt.figure(figsize=(16,5))
sns.distplot(daily_income_pop, label='Daily Income (Populasi)', color='blue')
plt.axvline(daily_income.income.mean(), color='red', linewidth=2, label='Daily Income (Rata-rata)')

plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%/Tingkat Kepercayaan 0,05')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%/Tingkat Kepercayaan 0,05')

plt.axvline(daily_income_pop.mean() + t_stat*daily_income_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis')
plt.axvline(daily_income_pop.mean() - t_stat*daily_income_pop.std(), color='black', linestyle='dashed', linewidth=2)
plt.legend()
```

### Dua Sampel Uji Hipotesis Dua Sisi Independen

Sekarang, kami ingin memeriksa, apakah rata-rata harian penjualan dua negara berbeda secara signifikan atau tidak menggunakan dua sampel uji dua arah yang independen. Kami akan memilih sampel Perancis dan Jerman.

seharusnya berbeda karena masing-masing negara memiliki pasarnya sendiri. Pertanyaannya apakah pasar prancis dan jerman itu signifikan?

```python
daily_france = data[data['Country']=='France'][['date','income']].groupby('date').sum()
daily_germany = data[data['Country']=='Germany'][['date','income']].groupby('date').sum()

print('Average sales of France a day: ${}'.format(np.round(daily_france.income.mean())))
print('Average sales of Germany a day: ${}'.format(np.round(daily_germany.income.mean())))
```

secara rata-rata memiliki selisih 5 dolar. pertanyaanya apakah 5 dolar sebuah kebetulan? atau signifikan?

Our hypothesis on this case:

**H0: μ_france = μ_germany** (apakah rata-rata prancis sama dengan jerman)

**H1: μ_france != μ_germany** (apakah rata-rata prancis tidak sama dengan jerman)

```python
#Kalau independent pakai stats.ttest_ind
t_stat, p_val = stats.ttest_ind(daily_france,daily_germany)
print('P-value:',p_val[0]) #the p-value isn't divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat[0])
```

hasilnya 0.4 > 0,05 maka gagal menolak ho. secara visual nilai p-vaue berada di dalam garis hitam dan tidak berada confidence thershols of 95%.

```python
france_pop = np.random.normal(daily_france.income.mean(),daily_france.income.std(),10000)
germany_pop = np.random.normal(daily_germany.income.mean(),daily_germany.income.std(),10000)

ci = stats.norm.interval(0.95, daily_france.income.mean(), daily_france.income.std())
plt.figure(figsize=(16,5))
sns.distplot(france_pop, label='Penjualan Rata-Rata Prancis Sehari *Pop',color='blue')
sns.distplot(germany_pop, label='Penjualan Rata-Rata Jerman Sehari *Pop',color='red')

plt.axvline(daily_france.income.mean(), color='blue', linewidth=2, label='Prancis rata-rata')
plt.axvline(daily_germany.income.mean(), color='red',  linewidth=2, label='Jerman rata-rata')

plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%/Tingkat Kepercayaan 0,05')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2)

plt.axvline(france_pop.mean()+t_stat[0]*france_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis')
plt.axvline(france_pop.mean()-t_stat[0]*france_pop.std(), color='black', linestyle='dashed', linewidth=2)

plt.legend()
```

Berdasarkan hasil di atas, kita dapat menyimpulkan bahwa **kita gagal menolak hipotesis nol** yang antara Perancis dan Jerman tidak berbeda secara signifikan dalam hal penjualan rata-rata per hari.

## ANOVA

ANOVA mirip dengan uji-t. Digunakan untuk menguji apakah lebih dari dua variabel berbeda nyata atau tidak. Jadi, kita akan menguji apakah rata-rata penjualan harian IEV,N8U, dan U5F berbeda nyata atau tidak.

```python
scanner_data = pd.read_csv('scanner_data.csv')
scanner_data
```

```python
scanner_data.groupby('Date').sum()
```

```python
#Melihat rata-rata 
IEV_quantity = scanner_data[scanner_data.SKU_Category == 'IEV'].groupby('Date').sum()['Quantity'] #Dikelompokkan groupby data jumlah date dan quantity nya
N8U_quantity = scanner_data[scanner_data.SKU_Category == 'N8U'].groupby('Date').sum()['Quantity']
U5F_quantity = scanner_data[scanner_data.SKU_Category == 'U5F'].groupby('Date').sum()['Quantity']

print("Daily Average of IEV",IEV_quantity.mean())
print("Daily Average of N8U",N8U_quantity.mean())
print("Daily Average of U5F",U5F_quantity.mean())
```

Tampaknya mereka berbeda secara signifikan karena rata-rata N8U adalah 33 dan sisanya adalah `22.

```python
# Anova pakai uji F
f_stat,p_value = stats.f_oneway(IEV_quantity, N8U_quantity, U5F_quantity)
print('P-value:',p_value)
```

**Karena nilai p di bawah 0,05, maka kami menolak Hipotesis Null.** Kami menyimpulkan bahwa perbedaan IEV, N8U, dan U5F signifikan secara statistik.

Hipotesis kami tentang kasus ini:

**H0: μ_gate30 = μ_gate40**

```python
print('Retention rate of gate 30 version:',gate_30.sum())
print('Retention rate of gate 40 version:',gate_40.sum())
#retention rate atau jumlah barisnya jika di jumlahkan bernilai 40000, makanya pakai 40000 sampel yang paling optimal
```

## Chi-Square Test

Uji chi-kuadrat digunakan untuk menguji independensi antara dua data kategorikal. Karena statistik menangani data numerik, kita perlu menghitung frekuensi setiap variabel dan disajikan dengan tabel kontingensi.

H0 : gender dan kesukaan belanja tidak ada hubungan

H1 : gender dan kesukaan belanja saling berhubungan

```python
chi_df=pd.read_csv('chi-test.csv')
chi_df
```

```python
#untuk meihat apakah gender ity mempengaruhi tingkat kesukaan dalam berbelanja
contingency_table=pd.crosstab(chi_df["Gender"],chi_df["Like Shopping?"]) #memakai fungsi crosstab pada pandas
contingency_table
```

```python
stat, p, dof, expected = stats.chi2_contingency(contingency_table)
print('p=%.3f' % (p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

H0 nya benar
