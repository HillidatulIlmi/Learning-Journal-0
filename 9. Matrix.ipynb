# Matrix

Definisi matrix adalah sekumpulan bilangan yang disusun berdasarkan baris dan kolom, serta ditempatkan di dalam tanda kurung. Matrix memiliki tipe data array 2 dimensi, terdiri dari kolom dan baris. 

Tipe data di aljabar linear ada empat, yaitu :
skalar, vektor, matrix dan tensor,

1.   Skalar atua bilangan, berupa nol dimensi/array 0 dimensi. misal bilangan 10
2.   Vektor, berupa bilangan satu dimensi atau array 1 dimensi, yang referentasinya adalah baris.
3.   Matrix, berupa bilangan dua dimensi arat array 2 dimensi.
4.   Tensor, memiliki n array dimensi, yang berisi beberapa matrix.

Jika dibandingkan dengan tipe data pandas. Tipe data pandas terdiri dari :
- series -> 1 dimensi entri data yang referantasinya adalah kolom
- data frame -> 2 dimensi data, hampir seperti matrix yang memiliki baris dan kolom

Trimminologi dimensi yaitu data dan refrentasi. Misalnya :
- data tabular, punya 2 kolom yaitu tinggi dan berat badan. misalnya ada data  (ilmi tinggi 164, berat 54) (tio tinggi 180, berat 70), maka dimesin data nya 2 dan refresentasi nya 2 dengan membentuk dua dimensi (x,y)
- kalau kita ingin menambahkan data baru yaitu umur (ilmi 24 dan tio 28), maka dimesi data nya 2 dan refresentasi nya jadi 3 dimensi, dengan membentuk 3 dimensi (x, y dan z)

3. Perkalian
- Perkalian matrix dengan skalar pakai simbol kali seperti biasa. C = skalar * matriks A

- Hadamard Product adalah operasi binear yang mengambil dua matriks berdimensi sama dan menghsikan matriks lain yang berdimensi sama dengan operan. Perkalian elemen  ways memakai simbol lingkaran bulat kecil. kalau pemograman pakai bintang *, C = matriks (m,n) dikali matriks (m,n).

- Perkalian matrix dengan matrix/matrix multipilcation adalah operasi binear yang menghasilkan matriks dari dua matriks. Untuk perkalian matriks, jumlah kolom pada matriks pertama sama dengan jumlah baris pada matriks kedua. Pakai simbol lingkaran bulat kecil (.), kalau pemograman pakai dot (@). Peraturan sebelum perkalian matrix dengan matrix :
    - jika matrix berukuran A(m,k) dan B(k,n). jika baris matrix A sama dengan kolom matrix B, maka bisa di lakukan operasi perkalian A(4,2) dan B(2,4) bisa di kalikan karena baris matrix A sama kolom matrix B
    - matrix tidak bersifar komutatif atau bisa di ubah. jika 7 x 3 = 3 x 7, kalau bilangan bisa terjadi/komutatif , tetapi kalau di matrix tidak bisa di operasikan

```python
# Perkalian matriks dengan matriks
H = np.array([[2,1,0],
              [0,2,2],
              [1,0,1],
              [3,0,1]
              ])
H # elemennya (4,3)
```

```python
J = np.array([[1,0],
              [0,4],
              [5,2]
              ])
J # elemennya (3,2)
```

```python
H @ J # perkalian dengan sheps yang berbeda 
# H ukuran nya (4,3) memiliki baris 4 dengan kolom 3
# J ukuran nya (3,2) memiliki baris 3 dengan kolom 2
# karena kolom matriks pertama sama dengan baris matriks kedua maka bisa di kalikan
```

```python
 J @ H # hasil nya akan error karena berbeda ukuran (n?,k),(k,m?)->(n?,m?) (size 4 is different from 2)
 # sehingga aturan komutatif tidak bisa 
```

2. Determinan

Mencari tau, apakah sebuah matrix memiliki invers.
jika hasil determinan nol, maka tidak memiliki matrix

```python
# Matrix determinant 3 X 3
from numpy import array
from numpy.linalg import det
# define matrix
F = array([[3,4,5],[7,7,12],[12,3,1]])
# calculate determinant
N = det(F)
print(N)
```

```python
# determinan 2 x 2
K = np.array([[2,1],
              [2,8]
              ])
K
```
