```python
i = 0

while i < 5:
  print(i)
  i = i + 1
```

```python
i = 1

while i <= 10:
  print(f'5 x {i} = {5*i}')
  i = i + 1
```

```python
# nested if

i = 1
j = 1

while i <= 10:
  while j <= 10:
    print(f'{i} x {j} = {i*j}')
    j = j + 1
  print('-'*10)
  i = i + 1
  print(f'nilai i = {i}')
  j = 1
```

### for loop

```python
cities = cities + ['Balik Papan', 'New York', 'Medan']
cities
```

```python
# for loop - cara biasa
for i in cities:
  print(i)
```

```python
# foor loop - menggunakan index
for idx in range(0,len(cities)):
  print(cities[idx])
```

```python
# while loop dengan hasil yang sama
i = 0

while i < len(cities):
  print(cities[i])
  i = i + 1
```

## Flow Control Statement

```python
# continue

for i in range(1,11):
  if i in [3, 7, 5, 6, 10]:
    continue
  print(i)
```

```python
# break

for i in range(1,11):
  print(i)
  if i in [7]:
    break
```

```python
dic = {'Rafif': 60, 'Adit': 75, 'Tio': 56, 'Surya': 80, 'Abi': 75}

# testing dict.items(): return key dan value
for i, n in dic.items():
  print(i)
  print(n)
  print('-'*10)
```

```python
# implementasi conditional statement dan looping
dic = {'Rafif': 60, 'Adit': 101, 'Tio': 56, 'Surya': 80, 'Abi': 75}

grade = {}
for nama, nilai in dic.items():
  if 0 <= nilai <= 100: 
    if 80 <= nilai <= 100:
      grade[nama] = 'A'
    elif 70 <= nilai < 80:
      grade[nama] = 'B'
    elif 60 <= nilai < 70:
      grade[nama] = 'C'
    elif 40 <= nilai < 60:
      grade[nama] = 'D'
    else:
      grade[nama] = 'E'
  else:
    grade[nama] = 'nilai diluar rentang'
```
